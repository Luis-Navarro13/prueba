```python
import matplotlib.pyplot as plt
```

```python
# Grafico de linea

x = [1, 2, 3, 4, 5]
y = [3, 5, 6, 7, 11]
```

```python
plt.plot(x,y)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Grafico de linea')
plt.show()
```

```python
# Histogramas
import numpy as np
datos = np.random.randn(1000)
```

```python
plt.hist(datos, bins = 30, color = 'skyblue', edgecolor = 'black')
plt.show()
```

```python
# Grafica de barras

categorias = ['x', 'y', 'z']
valores = [50, 60, 70]
```

```python
plt.bar(categorias, valores, color='green')
plt.xlabel('Categorias')
plt.ylabel('Valores')
plt.title('Grafico de barras')
plt.show()
```
